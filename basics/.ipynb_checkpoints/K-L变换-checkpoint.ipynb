```python
#K-L散度
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
W1=np.asmatrix([[0,0,0],[2,0,0],[2,0,1],[1,2,0]])
W2=np.asmatrix([[0,0,1],[0,1,0],[0,-2,1],[1,1,-2]])
#绘制散点图
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')

x1s=np.asarray(W1.T[0])
y1s=np.asarray(W1.T[1])
z1s=np.asarray(W1.T[2])

x2s=np.asarray(W2.T[0])
y2s=np.asarray(W2.T[1])
z2s=np.asarray(W2.T[2])

#使用K-L散度进行降维
#计算自相关矩阵R
R=0.5*0.25*(W1.T*W1+W2.T*W2)
#求R的特征值和特征向量
l,phi=np.linalg.eig(R)
print(l)
print(phi)
#降到2维
#抽取矩阵中的多行（类似切片）
a=phi[1:]
W12=W1*a.T
W22=W2*a.T
print('W12')
print(W12)
print('W22')
print(W22)
#降到1维
b=phi[2:]
print('W11')
print(W11)
print('W21')
print(W21)
W11=W1*b.T
W21=W2*b.T


ax.scatter(x1s,y1s,z1s,c='r')
ax.scatter(x2s,y2s,z2s,c='g')
plt.show()
```
